# Encoding Categorical Variables :

```python
le = LabelEncoder() # label encoder 
cendata['Income']=le.fit_transform(cendata['Income']) 
cendata['Sex']=le.fit_transform(cendata['Sex'])
#Converting 2 columns into binary 
cendata = pd.get_dummies(cendata,drop_first=True)
pd.set_option('display.max_columns',100)
```

```python
cendata.head()
```

Now our almost data values is 0 and 1 except few features like “‘Age’,’Fnlwgt’,’Education_num’,’Hours_per_week’” we can use standard scaler we and convert those features in same scale.



```python
scaler = StandardScaler()
train_col_sacle = cendata[['Age','Fnlwgt','Education_num','Hours_per_week']]
train_scaler_col = scaler.fit_transform(train_col_sacle)
train_scaler_col = pd.DataFrame(train_scaler_col,columns=train_col_sacle.columns)
cendata['Age']= train_scaler_col['Age']
cendata['Fnlwgt']= train_scaler_col['Fnlwgt']
cendata['Education_num']= train_scaler_col['Education_num']
cendata['Hours_per_week']= train_scaler_col['Hours_per_week']
```
